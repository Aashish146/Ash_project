```python
train_score=logr.predict_proba(x_train)[:,1]
real=y_train
```

```python
cutoffs=np.linspace(.01,.99,100)
```

```python
KS_all=[]
for cutoff in cutoffs:
    predicted=(train_score>cutoff).astype(int)
    TP=((predicted==1) & (real==1)).sum()
    TN=((predicted==0) & (real==0)).sum()
    FP=((predicted==1) & (real==0)).sum()
    FN=((predicted==0) & (real==1)).sum()
    
    P=TP+FN
    N=FP+TN
    
    KS=(TP/P)-(FP/N)
    
    KS_all.append(KS)
    
```

```python
 my_cutoff=cutoffs[KS_all==max(KS_all)][0]
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import Ridge, Lasso
```

```python
c_param=np.linspace(.01,99,100)
```

```python
rg_train.reset_index(inplace=True,drop=True)
x_train=rg_train.drop('Revenue.Grid',axis=1)
y_train=rg_train['Revenue.Grid']
```

```python
auc_score=[]
for c in c_param:
    logr=LogisticRegression(C=c,penalty='l1',class_weight='balanced',random_state=2)
    kf=KFold(n_splits=10)
    myauc=0
    
    for train,test in kf.split(x_train):
        logr.fit(x_train.loc[train],y_train[train])
        p=logr.predict(x_train.loc[test])
        myauc+=roc_auc_score(y_train[test],p)
    
    auc_10cv=myauc/10
    print(c,':',auc_10cv)
    auc_score.extend([auc_10cv])
    
```

```python
best_c=c_param[auc_score==max(auc_score)]
print('Value of C with max AUC score:',max(auc_score),' for 10 fold CV is :',best_c )
```
